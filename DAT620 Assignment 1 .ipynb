```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Add nodes (names from the Krackhardt kite diagram)
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]
G.add_nodes_from(nodes)

# Add edges based on the connections in the diagram
# (Andre connects to Beverly, Diane, Carol;
# Beverly to Diane and Ed;
# Diane to Fernando and Garth;
# Fernando to Heather and Ike;
# Garth to Heather;
# Heather to Ike and Jane;
# Ike to Jane)
edges = [
    ("Andre", "Beverly"), ("Andre", "Diane"), ("Andre", "Carol"),
    ("Beverly", "Diane"), ("Beverly", "Ed"),
    ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Fernando", "Heather"), ("Fernando", "Ike"),
    ("Garth", "Heather"),
    ("Heather", "Ike"), ("Heather", "Jane"),
    ("Ike", "Jane")
]
G.add_edges_from(edges)

```

```python
# Visualize the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1500, font_size=10, font_weight='bold', pos=nx.spring_layout(G))
plt.title("Krackhardt Kite Graph")
plt.show()

# Optional: Print basic graph info to verify
print("Nodes:", G.nodes())
print("Edges:", G.edges())
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

```python
# Define shells for the shell layout (group nodes by connectivity)
# Central nodes (Andre, Diane) in the inner shell, others in outer shell
shells = [["Andre", "Diane"], ["Beverly", "Carol", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]]
plt.figure(figsize=(8, 6))
pos = nx.shell_layout(G, shells)
nx.draw(G, pos, with_labels=True, node_color='lightyellow', node_size=1500, font_size=10, font_weight='bold', edge_color='gray')
plt.title("Krackhardt Kite Graph - Shell Layout")
plt.show()
```
