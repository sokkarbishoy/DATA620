### Davis Southern Women Dataset Analysis
#### Introduction
This Jupyter notebook analyzes the Davis Southern Women dataset, a classic bipartite network dataset from a 1941 study by Davis, Gardner, and Gardner. It records the attendance of 18 Southern women at 14 social events. The graph is bipartite, with women on one side and events (clubs) on the other.

#### Dataset Source

Built-in in NetworkX: nx.davis_southern_women_graph().
Original data: UCINET .dat file.
NetworkX example: Documentation.

#### Methodology: 
1. We load the bipartite graph using NetworkX and perform the following steps:
2. Graph Visualization and Biadjacency Matrix: Display the bipartite structure and attendance matrix to understand raw connections.
Projections: Create unweighted and weighted projections onto women and events to infer relationships based on shared attendances.
    - Unweighted: Binary connections (co-attended at least once).
    - Weighted: Strength based on number of shared events/attendees.
3. Analysis Techniques:
    - Compute degrees and centralities to identify central figures/events.
    - Detect communities using greedy modularity to find cliques or clusters.
4. Inferences: Draw sociological insights from patterns, such as social hierarchies or event groupings.

This approach allows us to transform attendance data into meaningful social network insights, revealing homophily, centrality, and community structures in 1930s Southern society.

```python
import networkx as nx
from networkx.algorithms import bipartite
from networkx.algorithms.community import greedy_modularity_communities
import matplotlib.pyplot as plt
import numpy as np
from scipy.sparse import csr_matrix  # For matrix display if needed

# Load the bipartite graph
G = nx.davis_southern_women_graph()
women = sorted(G.graph["top"])  # 18 women
clubs = sorted(G.graph["bottom"])  # 14 events (E1 to E14)

print(f"Number of women: {len(women)}")
print(f"Number of events: {len(clubs)}")
print(f"Number of edges (attendances): {G.number_of_edges()}")
print("\nWomen:", women)
print("Events:", clubs)
```

### Bipartite Graph Visualization
In this section, we visualize the bipartite graph to see direct connections between women and events. We also print the biadjacency matrix for a tabular view of attendances.

```python
# Layout for visualization
pos = nx.spring_layout(G, seed=45757)  

# Draw bipartite graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color=['lightblue' if node in women else 'lightgreen' for node in G.nodes()],
        node_size=500, font_size=8, font_weight='bold')
plt.title("Davis Southern Women Bipartite Graph")
plt.show()

# Biadjacency matrix (sparse print for brevity)
print("Biadjacency Matrix (Women x Events):")
biadj = bipartite.biadjacency_matrix(G, women, clubs)
print(biadj)
```

### Projections and Relationships Between Women
To infer relationships among women, we project the graph onto the women nodes. This creates a network where women are connected if they co-attended events. We compute degrees (unweighted for number of co-attendees, weighted for total shared events) and sort them to identify central figures. Community detection reveals social cliques.

```python
# Unweighted projection
W = bipartite.projected_graph(G, women)

# Weighted projection
W_weighted = bipartite.weighted_projected_graph(G, women)

print("# Co-attended events (unweighted degree), Woman")
for w in sorted(women, key=lambda x: W.degree(x), reverse=True):
    print(f"{W.degree(w):2d} {w}")

print("\n# Shared events (weighted degree), Woman")
for w in sorted(women, key=lambda x: W_weighted.degree(x, weight='weight'), reverse=True):
    print(f"{W_weighted.degree(w, weight='weight'):2d} {w}")
```

### Community Detection Among Women

```python
# Greedy modularity communities
communities_w = list(greedy_modularity_communities(W))
print("Women's Communities (based on co-attendance):")
for i, comm in enumerate(communities_w, 1):
    print(f"Community {i} ({len(comm)} women): {sorted(comm)}")
```

### Centrality in Women's Network

```python
centrality_w = nx.degree_centrality(W)
print("Degree Centrality (Women's Projection):")
for w in sorted(women, key=lambda x: centrality_w[x], reverse=True):
    print(f"{centrality_w[w]:.3f} {w}")
```

Inferences on Women Relationships

Central Women: High-degree women like Evelyn Jefferson and Theresa Anderson co-attend with nearly all others, forming the social core.
Communities: Two cliques suggest segregated groups: a large inner circle (12 women) and a smaller peripheral group (6 women), indicating homophily based on event choices.
Hierarchy: Central women act as bridges, connecting subgroups and influencing social dynamics.

Projections and Relationships Between Events
Similarly, we project onto events to find similarities based on shared attendees. This helps infer event types or popularity based on overlapping crowds.

```python
# Unweighted projection
C = bipartite.projected_graph(G, clubs)

# Weighted projection
C_weighted = bipartite.weighted_projected_graph(G, clubs)

print("# Shared women (unweighted degree), Event")
for c in sorted(clubs, key=lambda x: C.degree(x), reverse=True):
    print(f"{C.degree(c):2d} {c}")

print("\n# Overlapping attendees (weighted degree), Event")
for c in sorted(clubs, key=lambda x: C_weighted.degree(x, weight='weight'), reverse=True):
    print(f"{C_weighted.degree(c, weight='weight'):2d} {c}")
```

```python
# Greedy modularity communities
communities_c = list(greedy_modularity_communities(C))
print("Event Communities (based on shared attendees):")
for i, comm in enumerate(communities_c, 1):
    print(f"Community {i} ({len(comm)} events): {sorted(comm)}")
```

```python
centrality_c = nx.degree_centrality(C)
print("Degree Centrality (Events Projection):")
for c in sorted(clubs, key=lambda x: centrality_c[x], reverse=True):
    print(f"{centrality_c[c]:.3f} {c}")
```

Inferences on Event Relationships

Central Events: E6-E9 have high overlap, acting as integrative "hub" events.
Communities: Two clusters (E1-E7 vs. E8-E14) suggest distinct event types, possibly formal vs. informal gatherings.
Patterns: High-overlap events attract similar crowds, indicating themed or seasonal groupings.
